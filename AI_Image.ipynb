```python
import numpy as np
from PIL import Image
import tensorflow as tf
from tensorflow.keras.preprocessing import image

# Assuming 'cat.4018.jpg' is in the current directory
filename = 'cat.4018.jpg'
im = Image.open(filename)
im_resized = im.resize((224, 224))
resized_img = image.img_to_array(im_resized)

# Expand the dimensions to add the batch size dimension (1)
final_image = np.expand_dims(resized_img, axis=0)
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)

print(resized_img.shape)
```

```python
final_image.shape
```

```python
Predictions = mobile.predict(final_image)
print(Predictions )
```

```python
from tensorflow.keras.applications import imagenet_utils
Results = imagenet_utils.decode_predictions(Predictions)
print ( Results )
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.applications import imagenet_utils

# Assuming 'Predictions' contains the predictions
top_indices = np.argsort(Predictions[0])[::-1][:5]  # Get the top 5 indices
top_labels = imagenet_utils.decode_predictions(Predictions)[0]  # Get the top 5 labels

# Display the top 5 images
plt.figure(figsize=(15, 5))
for i, (label_id, label, prob) in enumerate(top_labels):
    plt.subplot(1, 5, i+1)
    plt.imshow(final_image[0] / 255.0)  # Assuming final_image is the preprocessed image
    plt.title(f"{label} ({prob:.2f})")
    plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
for i, (label_id, label, prob) in enumerate(top_labels):
    plt.subplot(1, 5, i+1)
    plt.imshow(final_image[0] / 255.0, interpolation='nearest')  # Set interpolation to 'nearest'
    plt.title(f"{label} ({prob:.2f})")
    plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
for i, (label_id, label, prob) in enumerate(top_labels):
    plt.subplot(1, 5, i+1)
    # Ensure that final_image is properly normalized
    plt.imshow(final_image[0])
    plt.title(f"{label} ({prob:.2f})")
    plt.axis('off')
plt.show()
```

```python
MobileNet_Two = tf.keras.applications.MobileNetV2()
predictions = MobileNet_Two.predict(final_image)
Results = imagenet_utils.decode_predictions(predictions)
print(Results)
```

```python
for result in Results[0]:
    label_id, label, prob = result
    print(f"{label}: {prob:.2f}")
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

# Display the images corresponding to the top predictions
plt.figure(figsize=(15, 5))
for i, (label_id, label, prob) in enumerate(Results[0][:5]):
    plt.subplot(1, 5, i+1)
    plt.imshow(final_image[0] / 255.0)
    plt.title(f"{label} ({prob:.2f})")
    plt.axis('off')
plt.show()
```
